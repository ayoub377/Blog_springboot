```python
# Import necessary libraries
import os
from pdf2image import convert_from_path

# Path to the folder containing the PDF files
pdf_folder = 'pdfs'  # replace with the path to your folder containing PDFs

# Folder to save the images
output_folder = '/content/images-converted-from-pdf'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Iterate over each file in the folder
for i,filename in enumerate(os.listdir(pdf_folder)):
    # Check if the file is a PDF
    if filename.endswith('.pdf'):
        pdf_path = os.path.join(pdf_folder, filename)

        # Convert PDF to list of images
        images = convert_from_path(pdf_path)

        # Save images directly to the output folder, prefixing with the PDF filename for distinction
        base_filename = os.path.splitext(filename)[0]
        for j, image in enumerate(images):
            image_path = os.path.join(output_folder, f's_{j}{i + 1}.png')
            image.save(image_path, 'PNG')

        print(f"Saved {len(images)} images for '{filename}' in '{output_folder}'")
```

```python
import os
import pickle
import zipfile
from paddleocr import PaddleOCR
```

```python
# Initialize the OCR instance
ocr = PaddleOCR(lang="fr")  # Need to run only once to download and load the model into memory

# Set the directory containing the images and where you want to save the .pkl files
image_dir = '/content/images_converted'
pickle_dir = '/content/pickle_data'  # Create a directory for saving .pkl files

# Create the pickle directory if it doesn't exist
os.makedirs(pickle_dir, exist_ok=True)

# Loop through the range of image numbers (400 to 423)
for image in os.listdir(image_dir):
    # Construct the image filename
    image_path = os.path.join(image_dir, image)
    if image.endswith(".png"):
        # Perform OCR on the image
        result = ocr.ocr(image_path, cls=False)

        # Define the pickle file name
        pickle_filename = f"{image}.pkl"
        pickle_path = os.path.join(pickle_dir, pickle_filename)

        # Save the OCR result data to a pickle file with the same name as the image
        with open(pickle_path, 'wb') as pickle_file:
            pickle.dump(result[0], pickle_file)

        print(f"OCR completed for {image} and saved as {pickle_filename}")

# Create a zip file containing all .pkl files
zip_filename = '/content/pickle_data.zip'

with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for root, dirs, files in os.walk(pickle_dir):
        for file in files:
            file_path = os.path.join(root, file)
            arcname = os.path.relpath(file_path, pickle_dir)
            zipf.write(file_path, arcname)

print(f"All .pkl files have been saved in '{pickle_dir}' and zipped as '{zip_filename}'")
```
